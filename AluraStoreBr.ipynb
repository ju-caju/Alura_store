### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()

lojas = {
    "Loja 1": loja,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}
```

### Inicializar dicionários para armazenar métricas

```python
faturamentos = {}
avaliacoes = {}
fretes = {}
categorias_populares = {}
mais_vendidos = {}
menos_vendidos = {}
```

### Calcular métricas para cada loja

```python
for nome, df in lojas.items():
    faturamentos[nome] = df['Preço'].sum()
    avaliacoes[nome] = df['Avaliação da compra'].mean()
    fretes[nome] = df['Frete'].mean()
    categorias_populares[nome] = df['Categoria do Produto'].value_counts().idxmax()
    mais_vendidos[nome] = df['Produto'].value_counts().idxmax()
    menos_vendidos[nome] = df['Produto'].value_counts().idxmin()
```

### Exibir resultados

```python
print("\n--- Faturamento Total ---")
for loja, valor in faturamentos.items():
    print(f"{loja}: R$ {valor:,.2f}")

print("\n--- Média de Avaliação ---")
for loja, valor in avaliacoes.items():
    print(f"{loja}: {valor:.2f}")

print("\n--- Custo Médio de Frete ---")
for loja, valor in fretes.items():
    print(f"{loja}: R$ {valor:.2f}")

print("\n--- Categoria Mais Popular ---")
for loja, categoria in categorias_populares.items():
    print(f"{loja}: {categoria}")

print("\n--- Produto Mais Vendido ---")
for loja, produto in mais_vendidos.items():
    print(f"{loja}: {produto}")

print("\n--- Produto Menos Vendido ---")
for loja, produto in menos_vendidos.items():
    print(f"{loja}: {produto}")
```

### Gráfico 1: Faturamento por loja

```python
plt.figure(figsize=(8, 5))
plt.bar(faturamentos.keys(), faturamentos.values(), color='skyblue')
plt.title("Faturamento Total por Loja")
plt.ylabel("R$ (Reais)")
plt.xlabel("Lojas")
plt.tight_layout()
plt.show()
```

### Gráfico 2: Média de avaliação por loja (pizza)

```python
plt.figure(figsize=(6, 6))
plt.pie(avaliacoes.values(), labels=avaliacoes.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Participação da Média de Avaliação por Loja")
plt.show()
```

### Gráfico 3: Custo médio de frete por loja (barra horizontal)

```python
plt.figure(figsize=(8, 5))
plt.barh(list(fretes.keys()), list(fretes.values()), color='orange')
plt.title("Custo Médio de Frete por Loja")
plt.xlabel("R$ (Reais)")
plt.tight_layout()
plt.show()
```
